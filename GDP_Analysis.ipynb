#### Excercise:
-  Calculate the average growth of states over the duration 2013-14, 2014-15 and 2015-16 by taking the mean of the row `(% Growth over previous year)`. Compare the calculated value and plot it for the states. Make appropriate transformations if necessary to plot the data. 
Report the average growth rates of the various states:
    -  Which states have been growing consistently fast, and which ones have been struggling?
    -  Curiosity exercise - what has been the average growth rate of your home state, and how does it compare to the national average over this duration?

```python
# Percentage Growth of States over the duration 2013-14, 2014-15 and 2015-16
mean_growth_percent = pd.DataFrame(GSDP_data.loc[(GSDP_data['Items  Description'] == '(% Growth over previous year)')
                                                       & ((GSDP_data.Duration == '2013-14') 
                                                          | (GSDP_data.Duration == '2014-15') 
                                                          | (GSDP_data.Duration == '2015-16'))])

# Alternative of the above 
#average_growth = GSDP_data.loc[(GSDP_data['Items  Description'] == '(% Growth over previous year)') & (GSDP_data.Duration != '2012-13')]

#Dropping the unwanted column "Items Description"
mean_growth_percent = mean_growth_percent.drop(['Items  Description'], axis=1)
mean_growth_percent
```

```python
#Restructuring the dataset
mean_growth_percent.set_index('Duration', inplace = True)
mean_growth_percent = mean_growth_percent.T

#Calculating the Average in descending order
mean_growth_percent['Average'] = round(mean_growth_percent.iloc[:,0:3].mean(axis=1),2)
mean_growth_percent = mean_growth_percent.sort_values(by= 'Average', ascending= False)
mean_growth_percent
```

```python
# Restructuring the dataset for plotting
mean_growth_percent.index.name = 'States'
mean_growth_percent.reset_index(inplace=True)
mean_growth_percent
```

```python
# ComparIng the calculated Average value and plotting it for the states
# Setting seaborn style
sns.set_style("whitegrid")

# Plotting the data into BarPlot 
plt.figure(1, figsize= (13,15))
plt.title('Average Growth Percentage of the States')
sns.barplot( x = 'Average', y = 'States', data = mean_growth_percent)
plt.show()
```

#### Report:
-  From the above Plot we can see that the North-East states like __`Mizoram`__, __`Tripura`__, __`Nagaland`__ are growing faster than the others, though we don't have the data for the period `2015-16` for them, while __`Goa`__, __`Meghalaya`__ states are struggling in it.

#### Exercise:
-  Categorise the states into four categories based on GDP per capita (`C1, C2, C3, C4` - C1 would have the highest per capita GDP, C4 the lowest). The quantile values are (0.20,0.5, 0.85, 1), i.e. the states lying between the 85th and the 100th percentile are in C1, those between 50th and 85th percentile are in C2 and so on.
-  For each category C1, C2, C3, C4:
    -  Find the top 3/4/5 sub-sectors (such as agriculture, forestry and fishing, crops, manufacturing etc.) [not primary, secondary and tertiary] which contribute to approx. 80% of the GSDP of each category

```python
#Define the categories C1, C2, C3, C4 based on GDP per Capita of States
master_df_2014_15['Quantile_Category'] = pd.qcut(master_df_2014_15[master_df_2014_15.Item == 'Per Capita GSDP (Rs.)']['2014-15'],[0,0.20, 0.5, 0.85, 1], labels=['C4','C3','C2','C1'])

# Define the Sub-Sectors according to the Categories C1, C2, C3, C4
df_by_categories = master_df_2014_15[master_df_2014_15['Item'].isin(['Agriculture, forestry and fishing',
                                         'Mining and quarrying',
                                         'Manufacturing',
                                         'Electricity, gas, water supply & other utility services',
                                         'Construction',
                                         'Trade, repair, hotels and restaurants',
                                         'Transport, storage, communication & services related to broadcasting',
                                         'Financial services',
                                         'Real estate, ownership of dwelling & professional services',
                                         'Public administration',
                                         'Other services',
                                         'Gross State Domestic Product',
                                         'Per Capita GSDP (Rs.)'])] 
df_by_categories
```

```python
# Filling the NaN columns with corresponding Quantile Categories of States
df_by_categories = df_by_categories.bfill()
df_by_categories = df_by_categories[df_by_categories.Item != 'Per Capita GSDP (Rs.)']
df_by_categories
```

```python
# Grouping by the Categories corresponding to the Sub-sectors
df_by_categories.set_index(['Item'],inplace= True)
df_by_categories = df_by_categories.pivot_table(values='2014-15', index='Quantile_Category', columns='Item', aggfunc='sum')
df_by_categories
```

```python
# Calculaing the Percentage Contribution of each Sub-Sectors
temp_df = df_by_categories.values
df_by_categories.iloc[:,0:12] = 100*(temp_df[:,0:12]/temp_df[:,4,None])
df_by_categories
```

```python
# Round-up the percentage values upto 2 decimal places & drop the GSDP column
df_by_categories = round(df_by_categories,2)
df_by_categories = df_by_categories.drop(['Gross State Domestic Product'], axis=1)

# Sorting the percentage values on Quantile Categories
df_by_categories = df_by_categories.sort_values(by=['C4','C3','C2','C1'], ascending=False, axis=1)
df_by_categories
```

```python
# Calculte which Sub-Sectors contribute approx. 80% of GSDP

# C4
sorted_df_c4 = df_by_categories.sort_values(by='C4', ascending=False, axis=1)
sorted_df_c4
```

```python
# C4 sum of percentage
c4_sum = sorted_df_c4.iloc[0,0:7].sum()
c4_sum
```

## Part-II: GDP and Education Drop-out Rates

```python
# loading Data II
GDP_Education = pd.read_csv("./Excel_files/Data-II-Download.csv")
GDP_Education
```

#### Exercise:
Analyse if there is any correlation of `GDP per capita` with `dropout rates in education` (`primary`, `upper primary` and `secondary`) for the year `2014-2015` for the states. Choose an appropriate plot to conduct this analysis.

```python
# Cleaning the data
GDP_Education = GDP_Education[['Level of Education - State','Primary - 2014-2015','Upper Primary - 2014-2015','Secondary - 2014-2015']]
GDP_Education = GDP_Education.rename(columns={'Level of Education - State' :'States'})
# Removing the Union Territories
GDP_Education = GDP_Education.drop([0,5,7,8,9,18,26])
GDP_Education
```

```python
# Renaming the column names by adding underscrore to it to match the State names with GSDP per capita dataframe
GDP_Education.set_index('States', inplace =True)
GDP_Education = GDP_Education.T
GDP_Education.columns = GDP_Education.columns.str.replace(' ', '_')
GDP_Education = GDP_Education.T
GDP_Education.reset_index(inplace=True)
GDP_Education
```

```python
# Data I-B
GSDP_per_capita
```

```python
# Merging Data I-B with Data II
merged_df = pd.merge(GDP_Education, GSDP_per_capita, how='inner', on= 'States')
merged_df = merged_df.drop(['Item'], axis =1)
merged_df = merged_df.rename(columns={'2014-15' :'GDP_per_capita_2014-15'})
merged_df
```

```python
# Required dataset to calculate correlation
before_corr_df = pd.DataFrame(merged_df.iloc[:,1:5])
before_corr_df
```

```python
# Calculating correlation 
correlation_df = before_corr_df.corr()
correlation_df = round(correlation_df, 3)
correlation_df
```

```python
# Plot correlation of GDP per capita with dropout rates in education (primary, upper primary and secondary)
plt.figure(figsize=(10,8))
plt.title('Correlation of GDP per capita with Dropout rates in education')
# heatmap
sns.heatmap(correlation_df.T, cmap="YlGnBu", annot=True)
plt.show()
```

#### Report:
The Blue boxes show the most correlated Dropout rates with respect to GDP per capita in education . 
So, __`Upper Primary (0.52)`__ dropout rate is highly correlated with __`Secondary (0.52)`__ dropout rate and also moderately correlated with __`Primary (0.45)`__.

Thus, to make educated skilled labour, one should consider the Drop-out rates from `Upper Primary` to `Secondary` and then also from `Primary` to `Upper Primary`.
